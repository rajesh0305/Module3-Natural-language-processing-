```python
import pandas as pd
bos= pd.DataFrame(boston.data)
print(bos.head())
```

```python
bos['price'] = boston.target

x = bos.drop('price',axis = 1)
y  = bos['price']
```

Adding 'price' column to the boston dataset and
x->all features data except the price features
y->only the price features

```python
# import sklearn
# x_train,x_test,y_train,y_test = sklearn.cross_validation.train_test_split(x,y)
# from sklearn.model_selection import train_test_split
# x_train,x_test,y_train,y_test = train_test_split(x,y)
import sklearn.model_selection as ms

x_train,x_test,y_train,y_test = ms.train_test_split(x,y,test_size=0.3,train_size=0.7,shuffle = True,random_state = 42)

print(x_train.shape)
print(x_test.shape)
print(y_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression


lm = LinearRegression()
lm.fit(x_train,y_train)
# or lm = LinearRegression().fit(x_train,y_train)

y_pred_lin = lm.predict(x_test)


plt.scatter(y_test,y_pred_lin)
plt.xlabel("Price:$y_i$")
plt.ylabel("Predicted price: $\hat{y}_i$")
plt.title("Linear Regression->price vs predicted price: $y_i$ vs $\hat{y}_i$")
plt.show()
```

```python
r2_score =lm.score(x_test,y_test)
print(r2_score*100,'%')
```

```python
def plotGraph(y_test,y_pred,regressorName):
    if max(y_test) >= max(y_pred):
        my_range = int(max(y_test))
    else:
        my_range = int(max(y_pred))
    plt.scatter(range(len(y_test)), y_test, color='blue',label = 'y_test')
    plt.scatter(range(len(y_pred)), y_pred, color='red',label='y_pred')
    plt.title(regressorName)
    plt.legend()
    plt.show()

    return

plotGraph(y_test, y_pred_lin, "Plotting Y_actual vs Y_predicted")
```

```python
errors = y_test-y_pred_lin

import seaborn as sns
import numpy as np
sns.set_style('whitegrid')
sns.kdeplot(np.array(errors),bw =0.1)
plt.show()
```

```python
sns.kdeplot(
    data=np.array(errors),
    cumulative=True, common_norm=False, common_grid=True,bw =0.1
)
```
