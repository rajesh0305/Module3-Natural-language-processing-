```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets, neighbors
```

```python
from matplotlib.colors import ListedColormap

def knn_comparision(data,n_neighbors = 15):
  '''
  This function finds K-NN and plots the data
  '''
  x =data[:,:2]
  y = data[:,2]

  #grid cell size
  h = 0.02
  cmap_light = ListedColormap(['#FFAAAA','#AAAAFF'])
  cmap_bold = ListedColormap(['#FF0000','#0000FF'])

  #the core classifier K-NN
  #clf->means classifier
  clf = neighbors.KNeighborsClassifier(n_neighbors)
  clf.fit(x,y)

  x_min,x_max = x[:,0].min()-1,x[:,0].max()+1
  y_min ,y_max =x[:,1].min()-1,x[:,1].max()+1

  #we create a mesh grid(x_min,y_min) to (x_max,y_max) with 0.02 grid space
  xx,yy = np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))

  #we predict the value (either 0 or 1 ) of each element in the grid
  z = clf.predict(np.c_[xx.ravel(),yy.ravel()])

  #xx.ravel( will give a flattern array)
  #np.c_ : translate slice objects to concatenation along the second axis
  #> np.c_[np.array([1,2,3]),np.array([4,5,6])]
  #> array([[1,4],
  #         [2,5],
  #         [3,6]]) (source : np.c_ documentation)
  z = z.reshape(xx.shape)

  #pcolormesh will plot the (xx,yy) grid with colors according to the value
  #it look like decision boundary
  plt.figure()
  plt.pcolormesh(xx,yy,z,cmap = cmap_light)

  #scatter plot of the given points
  plt.scatter(x[:,0],x[:,1],c =y,cmap = cmap_bold)

  #defining scale on both axises
  plt.xlim(xx.min(),xx.max())
  plt.ylim(yy.min(),yy.max())

  #set the title
  plt.title('K-value='+str(n_neighbors))
  plt.show()




```

```python
data = np.genfromtxt('/content/drive/MyDrive/demo_data_for_knn/6.overlap.csv', delimiter = ',')
knn_comparision(data,1)
knn_comparision(data,5)
knn_comparision(data,30)
knn_comparision(data,50)
```
